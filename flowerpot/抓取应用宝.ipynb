```python
from pprint import pprint
from requests_html import HTMLSession

CATEGORY_MAP = {
    "购物": 122,
    "阅读": 102,
    "新闻": 110,
    "视频": 103,
    "旅游": 108,
    "工具": 115,
    "社交": 106,
    "音乐": 101,
    "美化": 119,
    "摄影": 104,
    "理财": 114,
    "系统": 117,
    "生活": 107,
    "出行": 112,
    "安全": 118,
    "教育": 111,
    "健康": 109,
    "娱乐": 105,
    "儿童": 100,
    "办公": 113,
    "通讯": 116
}

session = HTMLSession()

for category, id in CATEGORY_MAP.items():
    url = "https://sj.qq.com/myapp/category.htm?orgame=1&categoryId=" + str(id)
    r = session.get(url)
    main = r.html.find('.main', first=True)
    links = main.links
    pprint(links)
    for link in links:
        pkg = link.replace("detail.htm?apkName=", "")
        with open(f'sj.qq.com/{category}', 'a+') as out:
            out.write(f'{pkg}\n')
```
